```python
# Задание task_04-1_01.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def load_data():

    input_line = input().strip()
    numbers = []
    for num_str in input_line.split():
        try:
            number = float(num_str)
            numbers.append(number)
        except ValueError:
            # Если встречается некорректное значение, вызываем исключение
            raise ValueError(f"Некорректное число: '{num_str}'")
    return numbers


def save_data(values, filename):

    with open(filename, 'w', encoding='utf-8') as file:
        for value in values:
            file.write(f"{value}\n")


try:
    filename = input("Введите имя файла: ").strip()
    values = load_data()
    save_data(values, filename)
    print(f"Данные успешно записаны в файл '{filename}'")

except ValueError as err:
    print("Числа были введены с ошибкой.")
    print("Описание:", err)

except (OSError, IOError) as err:
    print("Ошибка при сохранении файла с данными:", err)

except Exception as err:
    print("Произошла ошибка!")
    print("Тип:", type(err))
    print("Описание:", err)
```

```python
# Задание task_04-1_02.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def load_numbers(filename):

    numbers = []
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if line:  # Пропускаем пустые строки
                try:
                    number = float(line)
                    numbers.append(number)
                except ValueError:
                    # Пропускаем строки, которые нельзя преобразовать в число
                    continue
    return numbers

def append_results(filename, total_sum, maximum):

    with open(filename, 'a', encoding='utf-8') as file:
        file.write(f"\n{total_sum}\n{maximum}")

def main():

    try:
        filename = input("Введите имя файла: ").strip()

        # Загружаем числа из файла
        numbers = load_numbers(filename)

        if not numbers:
            print("Файл не содержит чисел для обработки.")
            return

        # Вычисляем сумму и максимум
        total_sum = sum(numbers)
        maximum = max(numbers)

        # Дописываем результаты в файл
        append_results(filename, total_sum, maximum)

        print(f"Результаты записаны в файл '{filename}'")
        print(f"Сумма: {total_sum}")
        print(f"Максимум: {maximum}")
        print(f"Обработано чисел: {len(numbers)}")

    except FileNotFoundError:
        print("Ошибка: файл не найден.")
    except PermissionError:
        print("Ошибка: нет прав доступа к файлу.")
    except (OSError, IOError) as err:
        print(f"Ошибка при работе с файлом: {err}")
    except Exception as err:
        print(f"Произошла непредвиденная ошибка: {err}")

if __name__ == "__main__":
    main()
```

```python
# Задание task_04-1_03.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def load_numbers(filename):

    numbers = []
    invalid_lines = []

    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line_num, line in enumerate(file, 1):
                line = line.strip()
                if line:  # Пропускаем пустые строки
                    try:
                        number = float(line)
                        numbers.append(number)
                    except ValueError:
                        # Сохраняем некорректные строки
                        invalid_lines.append(f"Строка {line_num}: '{line}'")
    except FileNotFoundError:
        print("Файл не найден. Будет создан новый файл.")
    except (OSError, IOError) as err:
        print(f"Ошибка при чтении файла: {err}")
        print("Продолжаем работу с доступными данными.")

    return numbers, invalid_lines

def append_results(filename, total_sum, maximum, invalid_lines):

    try:
        with open(filename, 'a', encoding='utf-8') as file:
            file.write(f"\n{total_sum}\n{maximum}")
    except (OSError, IOError) as err:
        print(f"Ошибка при записи в файл: {err}")
        return False
    return True

def main():
    filename = input("Введите имя файла: ").strip()

    # Загружаем числа из файла (функция сама обрабатывает ошибки чтения)
    numbers, invalid_lines = load_numbers(filename)

    # Выводим информацию о некорректных строках
    if invalid_lines:
        print("Обнаружены некорректные строки (игнорируются):")
        for line_info in invalid_lines:
            print(f"  {line_info}")

    # Вычисляем сумму и максимум (даже если файл пустой или не существует)
    if numbers:
        total_sum = sum(numbers)
        maximum = max(numbers)

        print(f"Обработано чисел: {len(numbers)}")
        print(f"Сумма: {total_sum}")
        print(f"Максимум: {maximum}")

        # Пытаемся записать результаты в файл
        if append_results(filename, total_sum, maximum, invalid_lines):
            print(f"Результаты записаны в файл '{filename}'")
        else:
            print("Не удалось записать результаты в файл")
    else:
        print("Нет чисел для обработки.")
        total_sum = 0
        maximum = None

        # Все равно пытаемся записать результаты (0 и None)
        if append_results(filename, total_sum, maximum, invalid_lines):
            print(f"Результаты записаны в файл '{filename}'")

if __name__ == "__main__":
    main()
```

```python
# Задание task_04-1_04.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

import os

def load_text(filename):
    """Загрузить стихотворение из файла 'filename'."""
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            return file.read()
    except UnicodeDecodeError:
        # Если UTF-8 не работает, пробуем другие кодировки
        with open(filename, 'r', encoding='cp1251') as file:
            return file.read()

def letters_stats(text):
    """Вернуть словарь, содержащий количество слов, начинающихся на гласную/согласную букву."""
    # Определяем множества гласных и согласных букв
    vowels = set('аеёиоуыэюяaeiou')
    consonants = set('бвгджзйклмнпрстфхцчшщbcdfghjklmnpqrstvwxyz')

    vowel_count = 0
    consonant_count = 0

    # Разбиваем текст на слова и обрабатываем каждое слово
    for word in text.split():
        # Очищаем слово от знаков препинания и приводим к нижнему регистру
        clean_word = word.strip('.,!?;:"()«»—').lower()

        if not clean_word:
            continue

        first_letter = clean_word[0]

        if first_letter in vowels:
            vowel_count += 1
        elif first_letter in consonants:
            consonant_count += 1

    return {
        "начинающихся на гласную": vowel_count,
        "начинающихся на согласную": consonant_count
    }

def create_sample_poem_file():
    """Создать файл с примером стихотворения, если он не существует."""
    sample_poem = """О сколько нам открытий чудных
Готовят просвещенья дух
И опыт, сын ошибок трудных,
И гений, парадоксов друг,
И случай, бог изобретатель..."""

    with open('poem.txt', 'w', encoding='utf-8') as f:
        f.write(sample_poem)
    print("Создан файл 'poem.txt' с примером стихотворения")

# Основная программа
def main():
    filename = input("Введите имя файла: ")

    # Проверяем существование файла
    if not os.path.exists(filename):
        print(f"Файл '{filename}' не найден!")
        print("Проверьте:")
        print("1. Правильно ли указано имя файла")
        print("2. Находится ли файл в той же папке, что и программа")
        print("3. Полный путь к файлу, если он в другой папке")

        create_choice = input("\nСоздать файл 'poem.txt' с примером стихотворения? (да/нет): ").lower()
        if create_choice in ['да', 'д', 'yes', 'y']:
            create_sample_poem_file()
            filename = 'poem.txt'
        else:
            print("Программа завершена.")
            return

    try:
        # Загружаем и обрабатываем текст
        text = load_text(filename)

        # Выводим стихотворение
        print("\n" + "=" * 50)
        print("СТИХОТВОРЕНИЕ:")
        print("=" * 50)
        print(text)
        print("=" * 50)

        # Получаем статистику
        stats = letters_stats(text)

        # Выводим подробную статистику
        print(f"\n СТАТИСТИКА:")
        print(f"   Слов, начинающихся на гласную: {stats['начинающихся на гласную']}")
        print(f"   Слов, начинающихся на согласную: {stats['начинающихся на согласную']}")

        # Определяем результат
        vowel_count = stats['начинающихся на гласную']
        consonant_count = stats['начинающихся на согласную']
        total_words = vowel_count + consonant_count

        print(f"\n ИТОГ:")
        print(f"   Всего проанализировано слов: {total_words}")

        if vowel_count > consonant_count:
            print("Слов, начинающихся на ГЛАСНУЮ букву, БОЛЬШЕ")
        elif vowel_count < consonant_count:
            print("Слов, начинающихся на СОГЛАСНУЮ букву, БОЛЬШЕ")
        else:
            print("Количество слов обеих категорий ОДИНАКОВО")

    except Exception as e:
        print(f"Ошибка при обработке файла: {e}")

# Запускаем программу
if __name__ == "__main__":
    main()
```

```python
# Задание task_04-1_05.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

class NoSuchSeatError(Exception):
    def __init__(self, message):
        super().__init__(message)


def load_seats(filename):

    seats = []
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            # Убираем пробелы и переводы строк, преобразуем каждый символ в int
            row = [int(char) for char in line.strip()]
            seats.append(row)
    return seats


def vacant_seats(seats):

    count = 0
    for row in seats:
        for seat in row:
            if seat == 0:  # 0 означает свободное место
                count += 1
    return count


def is_seat_vacant(seats, n_p, m_p):

    # Проверяем существование ряда
    if n_p < 0 or n_p >= len(seats):
        raise NoSuchSeatError(f"Ряда №{n_p + 1} не существует. Всего рядов: {len(seats)}")

    # Проверяем существование места в ряду
    if m_p < 0 or m_p >= len(seats[n_p]):
        raise NoSuchSeatError(f"Места №{m_p + 1} не существует в ряду №{n_p + 1}. Всего мест в этом ряду: {len(seats[n_p])}")

    return seats[n_p][m_p] == 0


def display_hall(seats):
    """Отобразить схему зала для наглядности."""
    print("\n" + "=" * 50)
    print("СХЕМА ЗАЛА (0 - свободно, 1 - занято):")
    print("=" * 50)
    for i, row in enumerate(seats):
        print(f"Ряд {i + 1}: {' '.join(map(str, row))}")
    print("=" * 50)


# Основная программа с обработкой исключений
try:
    filename = input("Введите имя файла: ")
    seats = load_seats(filename)

    # Показываем схему зала
    display_hall(seats)

    # Подсчитываем свободные места
    count = vacant_seats(seats)
    print(f"Всего свободных мест: {count}")

    # Ввод значений для поиска
    user_input = input("Введите ряд и место через пробел: ").split()

    if len(user_input) != 2:
        print("Ошибка: нужно ввести два числа через пробел!")
    else:
        n_p, m_p = [int(item) for item in user_input]

        # Проверяем корректность ввода (пользователь вводит с 1)
        if n_p < 1 or m_p < 1:
            print("Ошибка: номера ряда и места должны быть положительными числами!")
        else:
            try:
                is_vacant = is_seat_vacant(seats, n_p - 1, m_p - 1)
                status = "свободно" if is_vacant else "занято"
                print(f"Место в ряду {n_p}, место {m_p}: {status}")

            except NoSuchSeatError as e:
                print(f"{e}")

except FileNotFoundError:
    print(f"Ошибка: Файл '{filename}' не найден!")
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлу '{filename}'!")
except ValueError as e:
    print(f"Ошибка: Неверный формат данных в файле или при вводе!")
    print(f"   Подробнее: {e}")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")


# Дополнительно: интерактивный режим проверки мест
def interactive_mode(seats):
    """Режим многократной проверки мест."""
    print("\n" + "=" * 50)
    print("ИНТЕРАКТИВНЫЙ РЕЖИМ ПРОВЕРКИ МЕСТ")
    print("(для выхода введите 'exit')")
    print("=" * 50)

    while True:
        user_input = input("\nВведите ряд и место через пробел: ").strip().lower()

        if user_input in ['exit', 'выход', 'quit']:
            print("Завершение работы...")
            break

        parts = user_input.split()
        if len(parts) != 2:
            print("Ошибка: нужно ввести два числа через пробел!")
            continue

        try:
            n_p, m_p = [int(item) for item in parts]

            if n_p < 1 or m_p < 1:
                print("Ошибка: номера ряда и места должны быть положительными числами!")
                continue

            is_vacant = is_seat_vacant(seats, n_p - 1, m_p - 1)
            status = "свободно" if is_vacant else "занято"
            print(f"Место в ряду {n_p}, место {m_p}: {status}")

        except NoSuchSeatError as e:
            print(f"{e}")
        except ValueError:
            print("Ошибка: введите числа!")


# Запуск интерактивного режима (по желанию)
try:
    run_interactive = input("\nЗапустить интерактивный режим проверки? (да/нет): ").lower()
    if run_interactive in ['да', 'д', 'yes', 'y']:
        interactive_mode(seats)
except:
    pass
```

```python
# Задание task_04-1_06.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

CORRUPTED = -1

parties = {
    1: "Партия №1", 2: "Партия №2", 3: "Партия №3",
    4: "Партия №4", 5: "Партия №5", CORRUPTED: "Испорчено"
}


def load_votes(filename):

    votes = []
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()
        # Разбиваем содержимое на отдельные значения
        for vote_str in content.split():
            try:
                vote = int(vote_str)
                votes.append(vote)
            except ValueError:
                # Если значение не преобразуется в int, считаем бланк испорченным
                votes.append(CORRUPTED)
    return votes


def parties_votes(parties, votes):

    results = {}

    # Инициализируем счетчики для всех партий и испорченных бланков
    for party_name in parties.values():
        results[party_name] = 0

    # Подсчитываем голоса
    for vote in votes:
        if vote in parties and vote != CORRUPTED:
            party_name = parties[vote]
            results[party_name] += 1
        else:
            # Все остальные случаи (неизвестные партии, испорченные бланки)
            results[parties[CORRUPTED]] += 1

    return results


def print_results(votes_for_p):

    # Удаляем испорченные бланки из результатов для сортировки
    sorted_parties = []
    for party_name, votes_count in votes_for_p.items():
        if party_name != "Испорчено":
            sorted_parties.append((party_name, votes_count))

    # Сортируем партии по убыванию количества голосов
    sorted_parties.sort(key=lambda x: x[1], reverse=True)

    # Вычисляем общее количество действительных голосов (без испорченных)
    total_valid_votes = sum(votes_count for party_name, votes_count in sorted_parties)
    total_all_votes = total_valid_votes + votes_for_p.get("Испорчено", 0)

    # Выводим результаты для каждой партии
    for i, (party_name, votes_count) in enumerate(sorted_parties, 1):
        if total_valid_votes > 0:
            percentage = (votes_count / total_valid_votes) * 100
        else:
            percentage = 0.0

        # Форматируем вывод с выравниванием по правому краю
        print(f"{i}. {party_name} | {votes_count:>4} | {percentage:>6.2f}%")

    # Выводим информацию об испорченных бланках
    corrupted_count = votes_for_p.get("Испорчено", 0)
    if total_all_votes > 0:
        corrupted_percentage = (corrupted_count / total_all_votes) * 100
    else:
        corrupted_percentage = 0.0

    print(f"\nИспорченных бланков: {corrupted_count} ({corrupted_percentage:.2f}%)")
    print(f"Всего голосов: {total_all_votes}")


# Основная программа с обработкой исключений
try:
    filename = input("Введите имя файла: ")

    votes = load_votes(filename)
    votes_for_p = parties_votes(parties, votes)

    print("\n" + "=" * 50)
    print("РЕЗУЛЬТАТЫ ВЫБОРОВ")
    print("=" * 50)
    print_results(votes_for_p)

except FileNotFoundError:
    print(f"Ошибка: Файл '{filename}' не найден!")
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлу '{filename}'!")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
# Задание task_04-1_07.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

import csv


class NoSuchCountryError(Exception):
    def __init__(self, message):
        super().__init__(message)


class IllegalArgumentError(ValueError):
    pass


def load_data(filename):

    data = []
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            country_name = row['Country Name']
            gdp_value = row['2016']  # Используем данные за 2016 год

            # Пропускаем строки с пустыми значениями ВВП
            if gdp_value and gdp_value.strip():
                try:
                    gdp = float(gdp_value)
                    data.append({
                        "name": country_name,
                        "gdp": gdp
                    })
                except ValueError:
                    # Пропускаем некорректные числовые значения
                    continue
    return data


def search(data, criteria):

    if not data:
        raise NoSuchCountryError("Нет данных для поиска")

    # Определение операции
    if criteria == "-max-":
        # Находим страну с максимальным ВВП
        max_country = max(data, key=lambda x: x["gdp"])
        return max_country

    elif criteria == "-min-":
        # Находим страну с минимальным ВВП
        min_country = min(data, key=lambda x: x["gdp"])
        return min_country

    else:
        # Поиск по названию страны
        for country in data:
            if country["name"] == criteria:
                return country
        raise NoSuchCountryError(f"Страна '{criteria}' не найдена в данных")


def save_data(filename, data, criteria):

    try:
        method, value_str = criteria.split("=")
        filtered_data = []

        # Определение операции
        if method == "top":
            value = int(value_str)
            if value <= 0:
                raise IllegalArgumentError("Значение X должно быть > 0")
            # Сортируем по убыванию ВВП и берем первые X стран
            sorted_data = sorted(data, key=lambda x: x["gdp"], reverse=True)
            filtered_data = sorted_data[:value]

        elif method == "tail":
            value = int(value_str)
            if value <= 0:
                raise IllegalArgumentError("Значение X должно быть > 0")
            # Сортируем по возрастанию ВВП и берем первые X стран
            sorted_data = sorted(data, key=lambda x: x["gdp"])
            filtered_data = sorted_data[:value]

        elif method == "greater":
            value = float(value_str)
            # Фильтруем страны с ВВП больше value и сортируем по убыванию
            filtered_data = [country for country in data if country["gdp"] > value]
            filtered_data.sort(key=lambda x: x["gdp"], reverse=True)

        elif method == "less":
            value = float(value_str)
            # Фильтруем страны с ВВП меньше value и сортируем по возрастанию
            filtered_data = [country for country in data if country["gdp"] < value]
            filtered_data.sort(key=lambda x: x["gdp"])

        else:
            raise IllegalArgumentError(f"Неизвестный метод: {method}")

        # Сохранение в файл
        with open(filename, 'w', encoding='utf-8', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["name", "gdp"])
            writer.writeheader()
            for country in filtered_data:
                writer.writerow({
                    "name": country["name"],
                    "gdp": country["gdp"]
                })

    except ValueError as e:
        raise IllegalArgumentError(f"Некорректное значение в критерии: {e}")
    except Exception as err:
        raise IllegalArgumentError(
            "Значение параметра 'criteria' может быть "
            "одним из:\n"
            '- "top=X": первые X государств по ВВП на душу населения'
            ' (целое число > 0, по убыванию значения);\n'
            '- "tail=X": последние X государств по ВВП на душу населения'
            ' (целое число > 0, по возрастанию значения);\n'
            '- "greater=X": список государств с ВВП на душу населения, больше'
            ' чем X (вещ. число, по убыванию значения);\n'
            '- "less=X": список государств с ВВП на душу населения, меньше'
            ' чем X (вещ. число, по возрастанию значения).')


# Основная программа с обработкой исключений
try:
    filename = input("Введите имя файла для загрузки: ")
    save_filename = input("Введите имя файла для сохранения: ")

    # Загрузка данных
    data = load_data(filename)
    print(f"Загружено данных: {len(data)} стран")

    # Поиск стран
    print("\n" + "=" * 50)
    print("РЕЗУЛЬТАТЫ ПОИСКА:")
    print("=" * 50)

    # Страна с максимальным ВВП
    max_country = search(data, criteria="-max-")
    print(f"Страна с максимальным ВВП: {max_country['name']} - ${max_country['gdp']:,.2f}")

    # Страна с минимальным ВВП
    min_country = search(data, criteria="-min-")
    print(f"Страна с минимальным ВВП: {min_country['name']} - ${min_country['gdp']:,.2f}")

    # Поиск конкретной страны (например, России)
    try:
        russia = search(data, criteria="Russian Federation")
        print(f"Россия: {russia['name']} - ${russia['gdp']:,.2f}")
    except NoSuchCountryError:
        print("Россия не найдена в данных")

    # Сохранение топ-10 стран
    save_data(save_filename, data, criteria="top=10")
    print(f"\nТоп-10 стран сохранен в файл: {save_filename}")

    # Дополнительная информация
    print(f"\nДОПОЛНИТЕЛЬНАЯ ИНФОРМАЦИЯ:")
    print(f"Общее количество стран в данных: {len(data)}")
    avg_gdp = sum(country['gdp'] for country in data) / len(data)
    print(f"Средний ВВП на душу населения: ${avg_gdp:,.2f}")

except FileNotFoundError:
    print(f"Ошибка: Файл '{filename}' не найден!")
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлу '{filename}'!")
except NoSuchCountryError as e:
    print(f"Ошибка поиска страны: {e}")
except IllegalArgumentError as e:
    print(f"Ошибка в критерии: {e}")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
# Задание task_04-1_08.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

import csv
import json


class NoSuchYearError(Exception):
    def __init__(self, message):
        super().__init__(message)


def load_data(filename):

    data = []
    # Загрузка файла
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            data.append(row)

    # Преобразование строк в числа (кроме первого столбца и первой строки)
    for i in range(1, len(data)):
        for j in range(1, len(data[i])):
            if data[i][j].strip():  # Проверяем, что значение не пустое
                try:
                    data[i][j] = int(data[i][j])
                except ValueError:
                    # Если не удается преобразовать в int, оставляем как строку
                    pass

    return data


def export(filename, data, year_1, year_2):

    # Проверка параметров 'year_1' и 'year_2'
    headers = data[0]
    available_years = [int(year) for year in headers[1:-1] if str(year).isdigit()]

    if year_1 not in available_years:
        raise NoSuchYearError(f"Год {year_1} не найден в данных. Доступные годы: {available_years}")
    if year_2 not in available_years:
        raise NoSuchYearError(f"Год {year_2} не найден в данных. Доступные годы: {available_years}")

    if year_1 == year_2:
        raise NoSuchYearError("Годы для сравнения должны быть разными")

    # Находим индексы годов в заголовках
    year_1_index = headers.index(str(year_1))
    year_2_index = headers.index(str(year_2))

    # Разделяем данные на взрослых и детей
    adult_data = []
    children_data = []

    for i in range(1, len(data)):
        row = data[i]
        disease_class = row[0]
        age_group = row[-1]  # Последний столбец - возрастная группа

        # Получаем значения для обоих годов
        try:
            value_year_1 = int(row[year_1_index]) if row[year_1_index] else 0
            value_year_2 = int(row[year_2_index]) if row[year_2_index] else 0
        except (ValueError, TypeError):
            continue  # Пропускаем некорректные данные

        # Рассчитываем процентное изменение
        if value_year_1 != 0:
            percentage_change = ((value_year_2 - value_year_1) / value_year_1) * 100
        else:
            percentage_change = 0.0  # Если в первом году было 0, изменение 0%

        disease_info = {
            "класс болезни": disease_class,
            "значение": value_year_2,
            "изменение": round(percentage_change, 2)
        }

        if "дети" in age_group.lower() or "детск" in age_group.lower():
            children_data.append(disease_info)
        else:
            adult_data.append(disease_info)

    # Формирование списков
    # Сортируем по убыванию значений для второго года
    data_year_2_sorted_adult = sorted(
        [{"класс болезни": item["класс болезни"], "значение": item["значение"]}
         for item in adult_data],
        key=lambda x: x["значение"],
        reverse=True
    )

    data_year_2_sorted_children = sorted(
        [{"класс болезни": item["класс болезни"], "значение": item["значение"]}
         for item in children_data],
        key=lambda x: x["значение"],
        reverse=True
    )

    # Сортируем по убыванию процентных изменений
    data_changes_sorted_adult = sorted(
        [{"класс болезни": item["класс болезни"], "изменение": item["изменение"]}
         for item in adult_data],
        key=lambda x: x["изменение"],
        reverse=True
    )

    data_changes_sorted_children = sorted(
        [{"класс болезни": item["класс болезни"], "изменение": item["изменение"]}
         for item in children_data],
        key=lambda x: x["изменение"],
        reverse=True
    )

    res = {
        "Второй год": {
            "Взрослые": data_year_2_sorted_adult,
            "Дети": data_year_2_sorted_children
        },
        "Изменения": {
            "Взрослые": data_changes_sorted_adult,
            "Дети": data_changes_sorted_children
        },
        "Информация": {
            "Период сравнения": f"{year_1} → {year_2}",
            "Всего заболеваний у взрослых": len(adult_data),
            "Всего заболеваний у детей": len(children_data)
        }
    }

    # Сохранение в файл
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(res, file, ensure_ascii=False, indent=2)


# Основная программа с обработкой исключений
try:
    filename = input("Введите имя файла: ")
    export_filename = input("Введите имя файла для экспорта: ")

    # Загрузка данных
    data = load_data(filename)
    print(f"Данные успешно загружены. Строк: {len(data)}")

    # Показываем доступные годы
    headers = data[0]
    available_years = [year for year in headers[1:-1] if str(year).isdigit()]
    print(f"Доступные годы для анализа: {available_years}")

    # Ввод годов для сравнения
    year_input = input("Введите два года для сравнения через пробел: ").split()
    if len(year_input) != 2:
        print("Ошибка: нужно ввести ровно два года!")
    else:
        year_1, year_2 = int(year_input[0]), int(year_input[1])

        # Экспорт данных
        export(export_filename, data, year_1, year_2)
        print(f"Данные успешно экспортированы в файл: {export_filename}")
        print(f"Сравнение: {year_1} год → {year_2} год")

        # Дополнительная информация
        print(f"\nСтруктура JSON-файла:")
        print("- Второй год: данные за {year_2} год (отсортированы по убыванию)")
        print("- Изменения: процентные изменения {year_1}→{year_2} (отсортированы по убыванию)")
        print("- Разделы: Взрослые и Дети")

except FileNotFoundError:
    print(f"Ошибка: Файл '{filename}' не найден!")
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлу!")
except NoSuchYearError as e:
    print(f"Ошибка выбора года: {e}")
except ValueError as e:
    print(f"Ошибка ввода: введите корректные числовые значения для годов!")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
# Задание task_04-1_09.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

import json
import csv


class NoSuchFieldError(Exception):
    def __init__(self, message):
        super().__init__(message)


class IllegalArgumentError(ValueError):
    pass


def load_players(filename):

    # 1. Загрузить файл
    with open(filename, 'r', encoding='utf-8') as file:
        players = json.load(file)

    # 2. Добавить ключи
    for player in players:
        # Добавляем год рождения
        if "дата_рождения" in player:
            birth_date = player["дата_рождения"]
            player["год_рождения"] = int(birth_date.split("-")[0])

        # Добавляем ИМТ
        if "рост" in player and "вес" in player:
            height_m = player["рост"] / 100  # переводим см в метры
            weight_kg = player["вес"]
            player["ИМТ"] = weight_kg / (height_m ** 2)

            # Добавляем категорию ИМТ
            bmi = player["ИМТ"]
            if bmi <= 16:
                player["ИМТ_категория"] = "Выраженный дефицит массы тела"
            elif bmi <= 18.5:
                player["ИМТ_категория"] = "Недостаточная масса тела"
            elif bmi <= 24.99:
                player["ИМТ_категория"] = "Норма"
            elif bmi <= 30:
                player["ИМТ_категория"] = "Избыточная масса тела"
            elif bmi <= 35:
                player["ИМТ_категория"] = "Ожирение первой степени"
            elif bmi <= 40:
                player["ИМТ_категория"] = "Ожирение второй степени"
            else:
                player["ИМТ_категория"] = "Ожирение третьей степени"

    return players


def group_by(players, field):

    res = {}

    valid_fields = ["имя", "амплуа", "рука", "год_рождения", "клуб", "клуб_страна", "ИМТ_категория"]
    if field not in valid_fields:
        raise NoSuchFieldError(f"Поле '{field}' не поддерживается для группировки")

    for player in players:
        if field not in player:
            raise NoSuchFieldError(f"Нет данных для игрока по полю '{field}'")

        value = player[field]
        if value in res:
            res[value] += 1
        else:
            res[value] = 1

    return res


def save_group_data(filename, group_data, headers):

    if not isinstance(headers, list) or len(headers) != 2:
        raise IllegalArgumentError("Заголовки должны быть списком из 2 элементов")

    # Сортируем данные: сначала по убыванию количества, затем по возрастанию ключа
    sorted_items = sorted(group_data.items(), key=lambda x: (-x[1], x[0]))

    with open(filename, 'w', encoding='utf-8', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        for key, count in sorted_items:
            writer.writerow([key, count])


# Основная программа с обработкой исключений
try:
    filename = input("Введите имя JSON-файла с данными хоккеистов: ")
    save_filename = input("Введите имя CSV-файла для сохранения результатов: ")

    # Загрузка данных
    players = load_players(filename)
    print(f"Успешно загружено данных о {len(players)} хоккеистах")

    # Вывод доступных полей для группировки
    available_fields = ["имя", "амплуа", "рука", "год_рождения", "клуб", "клуб_страна", "ИМТ_категория"]
    print(f"\nДоступные поля для группировки: {', '.join(available_fields)}")

    # Выбор поля для группировки
    field = input("Введите поле для группировки: ")

    # Группировка данных
    group_data = group_by(players, field=field)

    # Определяем заголовки в зависимости от выбранного поля
    field_display_names = {
        "имя": "Имя",
        "амплуа": "Амплуа",
        "рука": "Рабочая рука",
        "год_рождения": "Год рождения",
        "клуб": "Клуб",
        "клуб_страна": "Страна клуба",
        "ИМТ_категория": "Категория ИМТ"
    }

    headers = [field_display_names.get(field, field.capitalize()), "Количество"]

    # Сохранение результатов
    save_group_data(save_filename, group_data, headers=headers)

    # Вывод статистики
    print(f"\nРезультаты группировки по полю '{field}':")
    print(f"Всего уникальных значений: {len(group_data)}")
    print(f"Данные сохранены в файл: {save_filename}")

    # Показываем топ-5 результатов
    sorted_items = sorted(group_data.items(), key=lambda x: (-x[1], x[0]))[:5]
    print("\nТоп-5 результатов:")
    for key, count in sorted_items:
        print(f"  {key}: {count}")

    # Дополнительная информация о хоккеистах
    print(f"\nОбщая информация о команде:")
    print(f"  Средний возраст: {sum(p.get('возраст', 0) for p in players) / len(players):.1f} лет")
    print(f"  Средний рост: {sum(p.get('рост', 0) for p in players) / len(players):.1f} см")
    print(f"  Средний вес: {sum(p.get('вес', 0) for p in players) / len(players):.1f} кг")

except FileNotFoundError:
    print(f"Ошибка: Файл '{filename}' не найден!")
except PermissionError:
    print(f"Ошибка: Нет прав доступа к файлу!")
except NoSuchFieldError as e:
    print(f"Ошибка поля: {e}")
except IllegalArgumentError as e:
    print(f"Ошибка аргументов: {e}")
except json.JSONDecodeError:
    print(f"Ошибка: Файл '{filename}' имеет некорректный JSON-формат!")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
# Задание task_04-1_10.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

import json


class IllegalArgumentError(ValueError):
    pass


def load_phonebook(filename):
    """Загрузить данные телефонной книги из json-файла 'filename'.

    Параметры:
        - filename (str): имя файла.

    Результат:
        - list of dict.

    Функция не обрабатывает исключения."""
    with open(filename, 'r', encoding='utf-8') as file:
        return json.load(file)


def save_phonebook(filename, phonebook):
    """Сохранить телефонную книгу 'phonebook' в json-файл 'filename'.

    Параметры:
        - filename (str): имя файла;
        - phonebook (list of dict): телефонная книга.

    Функция не обрабатывает исключения."""
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(phonebook, file, ensure_ascii=False, indent=2)


def search_by_name(phonebook, name):
    """Вернуть записи в телефонной книге, в имена которых входит 'name'
    (без учета регистра).

    Параметры:
        - phonebook (list of dict): телефонная книга;
        - name (str): имя контакта (не пустое).

    Результат:
        - list of dict.

    Исключения:
        - IllegalArgumentError: нарушение типа или содержимого 'name'.
    """
    if not isinstance(name, str) or not name.strip():
        raise IllegalArgumentError("Имя должно быть непустой строкой")

    name_lower = name.lower()
    results = []

    for contact in phonebook:
        if name_lower in contact.get("имя", "").lower():
            results.append(contact)

    return results


def search_by_phone(phonebook, phone_number):
    """Вернуть записи в телефонной книге, телефоны которых содержат
    'phone_number'.

    Параметры:
        - phonebook (list of dict): телефонная книга;
        - phone_number (str): номер телефона (не пустой).

    Результат:
        - list of dict.

    Исключения:
        - IllegalArgumentError: нарушение типа или содержимого 'phone_number'.
    """
    if not isinstance(phone_number, str) or not phone_number.strip():
        raise IllegalArgumentError("Номер телефона должен быть непустой строкой")

    results = []

    for contact in phonebook:
        phones = contact.get("телефоны", [])
        for phone in phones:
            if phone_number in phone.get("номер", ""):
                results.append(contact)
                break  # Не добавляем дубликаты одного контакта

    return results


def add_record(phonebook, name, phones):
    """Добавить в телефонную книгу запись с именем 'name' и списком
    телефонов 'phones'.

    Параметры:
        - phonebook (list of dict): телефонная книга;
        - name (str): имя (не пустое);
        - phones (list of dict): список телефонных номеров с описанием вида:
                                    {
                                        "описание": "мобильный",
                                        "номер": "+7111000043"
                                    }
                                 значения не должны быть пустыми.

    Результат:
        - phonebook (list of dict): телефонная книга.

    Исключения:
        - IllegalArgumentError:
          нарушение типа или содержимого 'name' или 'phones'.
    """
    if not isinstance(name, str) or not name.strip():
        raise IllegalArgumentError("Имя должно быть непустой строкой")

    if not isinstance(phones, list):
        raise IllegalArgumentError("Телефоны должны быть списком")

    for phone in phones:
        if not isinstance(phone, dict):
            raise IllegalArgumentError("Каждый телефон должен быть словарем")

        description = phone.get("описание", "")
        number = phone.get("номер", "")

        if not isinstance(description, str) or not description.strip():
            raise IllegalArgumentError("Описание телефона не должно быть пустым")

        if not isinstance(number, str) or not number.strip():
            raise IllegalArgumentError("Номер телефона не должен быть пустым")

    # Проверяем, нет ли уже контакта с таким именем
    existing_contact = None
    for contact in phonebook:
        if contact.get("имя", "").lower() == name.lower():
            existing_contact = contact
            break

    if existing_contact:
        # Добавляем телефоны к существующему контакту
        existing_contact["телефоны"].extend(phones)
    else:
        # Создаем новый контакт
        new_contact = {
            "имя": name,
            "телефоны": phones
        }
        phonebook.append(new_contact)

    return phonebook


def remove_name(phonebook, name):
    """Удалить из телефонной книги все контакты, содержащие 'name'
    (без учета регистра).

    Параметры:
        - phonebook (list of dict): телефонная книга;
        - name (str): имя (не пустое).

    Исключения:
        - IllegalArgumentError: нарушение типа или содержимого 'name'.
    """
    if not isinstance(name, str) or not name.strip():
        raise IllegalArgumentError("Имя должно быть непустой строкой")

    name_lower = name.lower()

    # Создаем новый список без удаляемых контактов
    new_phonebook = [
        contact for contact in phonebook
        if name_lower not in contact.get("имя", "").lower()
    ]

    # Очищаем исходный список и заполняем новыми данными
    phonebook.clear()
    phonebook.extend(new_phonebook)


def remove_phone(phonebook, phone_number):
    """Удалить из телефонной книги все телефоны, содержащие 'phone'.

    Параметры:
        - phonebook (list of dict): телефонная книга;
        - phone_number (str): номер телефона (не пустой).

    Результат:
        - phonebook (list of dict): телефонная книга.

    Исключения:
        - IllegalArgumentError: нарушение типа или содержимого 'phone_number'.
    """
    if not isinstance(phone_number, str) or not phone_number.strip():
        raise IllegalArgumentError("Номер телефона должен быть непустой строкой")

    for contact in phonebook:
        phones = contact.get("телефоны", [])
        # Оставляем только телефоны, которые НЕ содержат искомый номер
        contact["телефоны"] = [
            phone for phone in phones
            if phone_number not in phone.get("номер", "")
        ]

    return phonebook


def pretty_print(data):
    print(json.dumps(data, ensure_ascii=False, indent=2))


def display_contact(contact):
    """Красиво отобразить один контакт."""
    print(f"\nИмя: {contact['имя']}")
    print("Телефоны:")
    for i, phone in enumerate(contact.get('телефоны', []), 1):
        print(f"  {i}. {phone['описание']}: {phone['номер']}")


def interactive_mode():
    """Интерактивный режим работы с телефонной книгой."""
    filename = "phonebook.json"

    try:
        phonebook = load_phonebook(filename)
        print(f"Телефонная книга загружена. Контактов: {len(phonebook)}")
    except FileNotFoundError:
        print("Файл не найден. Создана новая телефонная книга.")
        phonebook = []
    except Exception as e:
        print(f"Ошибка загрузки: {e}")
        return

    while True:
        print("\n" + "=" * 50)
        print("ТЕЛЕФОННЫЙ СПРАВОЧНИК")
        print("=" * 50)
        print("1. Показать все контакты")
        print("2. Поиск по имени")
        print("3. Поиск по номеру")
        print("4. Добавить контакт")
        print("5. Удалить контакт")
        print("6. Удалить номер телефона")
        print("7. Сохранить и выйти")
        print("0. Выйти без сохранения")

        choice = input("\nВыберите действие: ").strip()

        if choice == "1":
            print(f"\nВсего контактов: {len(phonebook)}")
            for contact in phonebook:
                display_contact(contact)

        elif choice == "2":
            name = input("Введите имя для поиска: ").strip()
            try:
                results = search_by_name(phonebook, name)
                print(f"\nНайдено контактов: {len(results)}")
                for contact in results:
                    display_contact(contact)
            except IllegalArgumentError as e:
                print(f"Ошибка: {e}")

        elif choice == "3":
            phone = input("Введите номер для поиска: ").strip()
            try:
                results = search_by_phone(phonebook, phone)
                print(f"\nНайдено контактов: {len(results)}")
                for contact in results:
                    display_contact(contact)
            except IllegalArgumentError as e:
                print(f"Ошибка: {e}")

        elif choice == "4":
            name = input("Введите имя: ").strip()
            phones = []
            print("Введите телефоны (пустая строка для завершения):")
            while True:
                desc = input("Описание (мобильный, домашний и т.д.): ").strip()
                if not desc:
                    break
                number = input("Номер телефона: ").strip()
                if not number:
                    print("Номер не может быть пустым!")
                    continue
                phones.append({"описание": desc, "номер": number})

            try:
                phonebook = add_record(phonebook, name, phones)
                print("Контакт успешно добавлен!")
            except IllegalArgumentError as e:
                print(f"Ошибка: {e}")

        elif choice == "5":
            name = input("Введите имя для удаления: ").strip()
            try:
                before_count = len(phonebook)
                remove_name(phonebook, name)
                after_count = len(phonebook)
                print(f"Удалено контактов: {before_count - after_count}")
            except IllegalArgumentError as e:
                print(f"Ошибка: {e}")

        elif choice == "6":
            phone = input("Введите номер для удаления: ").strip()
            try:
                remove_phone(phonebook, phone)
                print("Номер(а) удалены из всех контактов!")
            except IllegalArgumentError as e:
                print(f"Ошибка: {e}")

        elif choice == "7":
            try:
                save_phonebook(filename, phonebook)
                print("Телефонная книга сохранена. До свидания!")
                break
            except Exception as e:
                print(f"Ошибка сохранения: {e}")

        elif choice == "0":
            print("Выход без сохранения.")
            break

        else:
            print("Неверный выбор!")


# Запуск интерактивного режима
if __name__ == "__main__":
    interactive_mode()

# Тестирование функций (раскомментируйте для проверки)
```
